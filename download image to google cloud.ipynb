```python

def download_image(url, file_name, category, split):
    # Send GET request
    response = requests.get(url)

    # Save the image
    folder_path = os.path.join("/content/drive/MyDrive/Traffy/{}".format(split), category)
    if response.status_code == 200:
        with open(os.path.join(folder_path, file_name), "wb") as f:
            f.write(response.content)
    else:
        print(response.status_code)
```
